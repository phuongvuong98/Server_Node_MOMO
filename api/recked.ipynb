```python
for key in file.keys():
    value = file[key]
    userStore = []
    for val in value:
        flag = 1
        storeList = transactions[transactions.tag == val].store_id.value_counts().keys()
        for store in storeList:
            if store in merchants.store_id.unique():
                userStore.append(merchants[merchants.store_id == store].index[0])
                break
    uDf = merchants.loc[userStore,:]
    reIndex = []
    for i in range(len(uDf)):
        reIndex.append("merchant"+str(i))
    uDf.index = reIndex
    fileName = "recommended\\"+key+".json"
    uDf.to_json(fileName,orient="index")
```
