```python
with open('data.json', 'r',encoding="utf-8") as myfile:
    data=myfile.read()
```

```python
newdat = data.replace("(","")
newdat = newdat.replace(")","")
newdat = newdat.replace("\n","")
newdat = newdat.replace("{","")
newdat = newdat.replace("}","")
newdat = newdat.replace("\"store\":","")
newdat = newdat.replace("\"link\":","")
newdat = newdat.replace("\"","")
newdat
```

```python
new2 = newdat.split(",")
```

```python
store = []
link = []
```

```python
for i in range(len(new2)):
  if i % 2 == 0:
    store.append(new2[i].strip())
  else:
    link.append(new2[i].strip())
```

```python
linkDict = dict()
```

```python
for i in range(len(store)):
  linkDict[store[i]]=link[i]
```

```python
def convert2link(tag):
  if tag in linkDict.keys():
    return linkDict[tag]
  else:
    return "https://localhost:3000/logo"
```

```python
merchants['link'] = merchants.tag.apply(convert2link)
```
